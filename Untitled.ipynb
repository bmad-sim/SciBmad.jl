```julia
tw = twiss(ring, GTPSA_descriptor=Descriptor([1,1,1,1,1,8], 8))
t = tw.table
```

```julia
deta_ddelta_x = map(x->x[[0,0,0,0,0,8]], t.orbit_x) # Multiply by 2! to get only derivative
deta_ddelta_y = map(x->x[[0,0,0,0,0,8]], t.orbit_y) # Multiply by 2! to get only derivative

f = Figure(fontsize=22, size=(700,250));
ax = Axis(f[1, 1], xlabel="s [m]")#, ylabel="Second-Order Dispersion [m]");
lines!(ax, t.s, deta_ddelta_x, label=L"\frac{1}{8!}\frac{\partial^8\eta_x}{\partial^8\delta}"); # L".." specifies a LaTeX string
lines!(ax, t.s, deta_ddelta_y, label=L"\frac{1}{8!}\frac{\partial^8\eta_y}{\partial^8\delta}");
Legend(f[1, 2], ax);
#axislegend(position=:rt);
display(f);
```

```julia
delta = (-1:0.01:1).*10^-2

Qx = tw.tunes[1]
Qy = tw.tunes[2]

# We can evaluate these Taylor series like any other function in Julia: Qx.(0,0,0,0,0,delta)
f = Figure(fontsize=22, size=(600,500));
ax = Axis(f[1, 1], xlabel=L"\delta", ylabel="Fractional Tunes") #, yticks=0.02:0.02:0.18);
lines!(ax, delta, Qx.(0,0,0,0,0,delta), label=L"Q_x"); # L".." specifies a LaTeX string
lines!(ax, delta, Qy.(0,0,0,0,0,delta), label=L"Q_y");
axislegend(position=:rt);
display(f);
```
