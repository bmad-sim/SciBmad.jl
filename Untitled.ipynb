```julia
using SciBmad, CairoMakie, LaTeXStrings
```

```julia
include("lattices/esr-v6.3.1.jl")
```

```julia
tw = twiss(ring, GTPSA_descriptor=Descriptor(6,2))
```

```julia
t = tw.table
set_theme!(theme_latexfonts())
deta_ddelta_x = map(x->x[[0,0,0,0,0,2]], t.orbit_x) # Multiply by 2! to get only derivative
deta_ddelta_y = map(x->x[[0,0,0,0,0,2]], t.orbit_y) # Multiply by 2! to get only derivative


f = Figure(fontsize=30, size=(800,300));
ax = Axis(f[1, 1], xlabel="s [m]");
lines!(ax, t.s, deta_ddelta_x, label=L"\frac{1}{2!}\frac{\partial^2x_{co}}{\partial\delta^2}"); # L".." specifies a LaTeX string
lines!(ax, t.s, deta_ddelta_y, label=L"\frac{1}{2!}\frac{\partial^2y_{co}}{\partial\delta^2}");
Legend(f[1,2], ax)
display(f);
```
