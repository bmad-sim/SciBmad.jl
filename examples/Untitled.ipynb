```julia
qf = Quadrupole(Kn1=0.5, L=0.5)
sf = Sextupole(Kn2=0.1, L=0.1)
d = Drift(L=1)
qd = Quadrupole(Kn1=-0.5, L=0.5)
sd = Sextupole(Kn2=-0.1, L=0.1)

fodo = Beamline([qf, sf, d, qd, sd, d], species_ref=Species("electron"), R_ref=-59.5287244902766);
```
