```julia
function plot_2nd_order(ring)
    # Second order
    tw2 = twiss(ring; GTPSA_descriptor=Descriptor(6, 2)) #, at=filter(x->x.kind=="Marker", ring.line))
    println("TUNES =================")
    println(tw2.tunes)
    println("=======================")
    t2 = tw2.table
    
    deta_ddelta_x = map(x->x[[0,0,0,0,0,2]], t2.orbit_x)
    deta_ddelta_y = map(x->x[[0,0,0,0,0,2]], t2.orbit_y)
    beta_1 = scalar.(t2.beta_1)
    beta_2 = scalar.(t2.beta_2)
    dbeta_ddelta_1 = map(x->x[[0,0,0,0,0,1]], t2.beta_1)
    dbeta_ddelta_2 = map(x->x[[0,0,0,0,0,1]], t2.beta_2)
    alpha_1 = scalar.(t2.alpha_1)
    alpha_2 = scalar.(t2.alpha_2)
    dalpha_ddelta_1 = map(x->x[[0,0,0,0,0,1]], t2.alpha_1)
    dalpha_ddelta_2 = map(x->x[[0,0,0,0,0,1]], t2.alpha_2)
    W_1 = @. sqrt((dalpha_ddelta_1-alpha_1/beta_1*dbeta_ddelta_1)^2 + (dbeta_ddelta_1/beta_1)^2)
    W_2 = @. sqrt((dalpha_ddelta_2-alpha_2/beta_2*dbeta_ddelta_2)^2 + (dbeta_ddelta_2/beta_2)^2)
    
    # Plot these guys
    f = Figure(fontsize=22, size=(700,2*250));
    ax1 = Axis(f[1, 1], xlabel="s [m]", ylabel="Montague Functions");
    ax2 = Axis(f[2, 1], xlabel="s [m]", ylabel="Second Order Dispersion [m]");
    
    lines!(ax1, t2.s, W_1, label=L"W_1"); 
    lines!(ax1, t2.s, W_2, label=L"W_2");
    lines!(ax2, t2.s, deta_ddelta_x, label=L"\frac{\partial\eta_x}{\partial\delta}"); 
    lines!(ax2, t2.s, deta_ddelta_y, label=L"\frac{\partial\eta_x}{\partial\delta}");
    
    axislegend(ax1, position=:rt);
    axislegend(ax2, position=:rt);
    axislegend(ax3, position=:rt);
    return f
end
display(plot_2nd_order(ring));
```

As you can see, the chromaticities are not exactly +1. Let's optimize the sextupoles to achieve that first.

To do this, we'll make all the focusing chromatic sextupoles be one knob, and all defocusing chromatic sextupoles another knob.

## Chromaticity Optimization

```julia
# First let's create a module to hold our knobs
module Controls end
```

```julia
# Make all focusing sextupoles one knob,
# defocusing other knob. We can do this using SciBmad's deferred expressions
sextupoles = filter(x->x.kind=="Sextupole", ring.line)
sfs = filter(x->x.Kn2 > 0, sextupoles)
sds = filter(x->x.Kn2 < 0, sextupoles)

# Define variables in Controls module:
Controls.ksf = first(sfs).Kn2 # Just set to current value
Controls.ksd = first(sds).Kn2 

# Now set the sextupoles as a knob:
foreach(x -> x.Kn2 = DefExpr(() -> Controls.ksf), sfs)
foreach(x -> x.Kn2 = DefExpr(() -> Controls.ksd), sds)
```

```julia
# Now use AD to make chromaticities +1

# Define function to return objective function (F) AND 
# gradient (G) IF PROVIDED:
function f_and_grad!(F, G, u, p)
    ring = p[1]
    control_mod = p[2]
    weight = p[3]
    chrom1_goal = p[4]
    chrom2_goal = p[5]

    if G !== nothing
        # 6 variables to 3rd order, 2 parameters to 1st order
        dchrom = Descriptor(6, 3, 2, 1)
        k = params(dchrom)
        control_mod.ksf = u[1] + k[1]
        control_mod.ksd = u[2] + k[2]
        tw = twiss(ring; GTPSA_descriptor=dchrom, at=[]) # at no elements, only tunes
        # Construct Jacobian
        chrom1 = tw.tunes[1][6]
        dchrom1_dksf = tw.tunes[1][[6=>1],params=[1=>1]]#this uses sparse monomial indexing, alternatively [[0,0,0,0,0,1,1,0]]
        dchrom1_dksd = tw.tunes[1][[6=>1],params=[2=>1]]# [[0,0,0,0,0,1,0,1]]
        #println(tw.tunes[1])
        chrom2 = tw.tunes[2][6]
        dchrom2_dksf = tw.tunes[2][[6=>1],params=[1=>1]]
        dchrom2_dksd = tw.tunes[2][[6=>1],params=[2=>1]]
        dQs_dV = tw.tunes[3][param=1]

        # Apply to gradient of sum of squares
        res = SA[chrom1-chrom1_goal, chrom2-chrom2_goal]
        J = SA[dchrom1_dksf dchrom1_dksd; dchrom2_dksf dchrom2_dksd]
        G .= 2*transpose(J)*res

        if F !== nothing
            return sum(abs2, res)
        end
    elseif F !== nothing # Just compute Qs, no derivative 
        # This is for gradient-free optimizers
        control_mod.ksf = u[1]
        control_mod.ksd = u[2]
        tw = twiss(ring; GTPSA_descriptor=Descriptor(6, 2), at=[]) # at no elements, only tunes
        chrom1 = tw.tunes[1][6]
        chrom2 = tw.tunes[2][6]
        res = SA[chrom1-chrom1_goal, chrom2-chrom2_goal] 
        return sum(abs2, res)
    end
end

# Optimize using Optim.jl
u0 = [2.5230882715314, -3.385942324738] # initial guess, use K2s already there
p = (ring, Controls, 1, 1.0, 1.0)

# Define closure with parameters
# Use Optim.NLSolversBase.only_fg! to specify f_and_gradient provided:
f = Optim.NLSolversBase.only_fg!((F,G,u)->f_and_grad!(F,G,u,p))

# Optimize
sol = Optim.optimize(f, u0, LBFGS())
```

```julia
# Now set the controls to the minimizer
Controls.ksf = sol.minimizer[1]
Controls.ksd = sol.minimizer[2]
```

```julia
# Check real quick:
twiss(ring; GTPSA_descriptor=Descriptor(6,2), at=[]).tunes
```

### Looks good! 

Now let's set up the sextupole families for W function and second order dispersion optimization.

# Sextupole Families

```julia
# First just make all focusing sextupoles one knob,
# defocusing other knob
sextupoles = filter(x->x.kind=="Sextupole", ring.line)
chrom_sextupoles = filter(x->x.Kn2 != 0, sextupoles)

arc1_sextupoles = filter(x->occursin(r"_1$", x.name), chrom_sextupoles)
arc3_sextupoles = filter(x->occursin(r"_3$", x.name), chrom_sextupoles)
arc5_sextupoles = filter(x->occursin(r"_5$", x.name), chrom_sextupoles)
arc7_sextupoles = filter(x->occursin(r"_7$", x.name), chrom_sextupoles)
arc9_sextupoles = filter(x->occursin(r"_9$", x.name), chrom_sextupoles)
arc11_sextupoles = filter(x->occursin(r"_11$", x.name), chrom_sextupoles)

# Arc sextupole layouts:
# 1:  D D F
# 3:  F D D
# 5:  D F F
# 7:  F F D
# 9:  D F F
# 11: F D D

# Easiest to separate F and D for each arc and apply according rules:
sfs_1 = filter(x->x.Kn2 > 0, arc1_sextupoles)
sds_1 = filter(x->x.Kn2 < 0, arc1_sextupoles)
sfs_3 = filter(x->x.Kn2 > 0, arc3_sextupoles)
sds_3 = filter(x->x.Kn2 < 0, arc3_sextupoles)
sfs_5 = filter(x->x.Kn2 > 0, arc5_sextupoles)
sds_5 = filter(x->x.Kn2 < 0, arc5_sextupoles)
sfs_7 = filter(x->x.Kn2 > 0, arc7_sextupoles)
sds_7 = filter(x->x.Kn2 < 0, arc7_sextupoles)
sfs_9 = filter(x->x.Kn2 > 0, arc9_sextupoles)
sds_9 = filter(x->x.Kn2 < 0, arc9_sextupoles)
sfs_11 = filter(x->x.Kn2 > 0, arc11_sextupoles)
sds_11 = filter(x->x.Kn2 < 0, arc11_sextupoles)

# One knob per plane per arc:
# Describes deviation of each family from 
# current values (which set chromaticity: [2.5230882715314, -3.385942324738]
Controls.dksf_1 = 0 
Controls.dksd_1 = 0 
Controls.dksf_3 = 0 
Controls.dksd_3 = 0 
Controls.dksf_5 = 0 
Controls.dksd_5 = 0 
Controls.dksf_7 = 0 
Controls.dksd_7 = 0 
Controls.dksf_9 = 0 
Controls.dksd_9 = 0 
Controls.dksf_11 = 0 
Controls.dksd_11 = 0 

# 1:  D D F
for i in 1:4:length(sds_1)-3
    sds_1[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_1)
    sds_1[i+1].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_1)
    sds_1[i+2].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_1)
    sds_1[i+3].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_1)
end

for i in 1:2:length(sfs_1)-1
    sfs_1[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_1)
    sfs_1[i+1].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_1)
end

# 3:  F D D
for i in 1:4:length(sds_3)-3
    sds_3[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_3)
    sds_3[i+1].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_3)
    sds_3[i+2].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_3)
    sds_3[i+3].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_3)
end

for i in 1:2:length(sfs_3)-1
    sfs_3[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_3)
    sfs_3[i+1].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_3)
end

# 5:  D F F
for i in 1:2:length(sds_5)-1
    sds_5[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_5)
    sds_5[i+1].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_5)
end

for i in 1:4:length(sfs_5)-3
    sfs_5[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_5)
    sfs_5[i+1].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_5)
    sfs_5[i+2].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_5)
    sfs_5[i+3].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_5)
end

# 7:  F F D
for i in 1:2:length(sds_7)-1
    sds_7[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_7)
    sds_7[i+1].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_7)
end

for i in 1:4:length(sfs_7)-3
    sfs_7[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_7)
    sfs_7[i+1].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_7)
    sfs_7[i+2].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_7)
    sfs_7[i+3].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_7)
end

# 9:  D F F
for i in 1:2:length(sds_9)-1
    sds_9[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_9)
    sds_9[i+1].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_9)
end

for i in 1:4:length(sfs_9)-3
    sfs_9[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_9)
    sfs_9[i+1].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_9)
    sfs_9[i+2].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_9)
    sfs_9[i+3].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_9)
end

# 11: F D D
for i in 1:4:length(sds_11)-3
    sds_11[i+0].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_11)
    sds_11[i+1].Kn2 = DefExpr(()->Controls.ksd+Controls.dksd_11)
    sds_11[i+2].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_11)
    sds_11[i+3].Kn2 = DefExpr(()->Controls.ksd-Controls.dksd_11)
end

for i in 1:2:length(sfs_11)-1
    sfs_11[i+0].Kn2 = DefExpr(()->Controls.ksf+Controls.dksf_11)
    sfs_11[i+1].Kn2 = DefExpr(()->Controls.ksf-Controls.dksf_11)
end
```

```julia
# Sanity check:
plot_2nd_order(ring)
```

```julia
# Let's adjust the controls and see what happens
Controls.dksf_1 = 0
Controls.dksd_1 = 0 
Controls.dksf_3 = -.2 
Controls.dksd_3 = 0 
Controls.dksf_5 = -1
Controls.dksd_5 = -1
Controls.dksf_7 = 1.6
Controls.dksd_7 = -2
Controls.dksf_9 = -0.3
Controls.dksd_9 = -1  
Controls.dksf_11 = 0 
Controls.dksd_11 = 0 
plot_2nd_order(ring)
```
